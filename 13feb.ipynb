Q2. Write a python program to print Python Exception Hierarchy.

```python
# Python program to print the Exception Hierarchy

def print_exception_hierarchy(exception, indent=0):
    for Hierarchy in exception:
        print(' ' * indent + Hierarchy.__name__)
        print_exception_hierarchy(Hierarchy.__subclasses__(), indent + 2)
        
print_exception_hierarchy([Exception])
```
